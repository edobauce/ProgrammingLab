### **1. Indicizzazione booleana**

1. Creare un nuovo vettore con tutti i numeri primi tra 0 e 20. I numeri primi compresi tra 0 e 20 sono: **2, 3, 5, 7, 11, 13, 17, 19**.

2. Utilizza un test logico (ad esempio `>`, `<`, `==`, `!=`) per **selezionare tutte le voci del vettore maggiori di 10**.

3. Utilizza un test logico per selezionare **tutti i numeri primi pari** dalla lista.  


```python
import numpy as np

prime_array = np.array([2,3,5,7,11,13,17,19])

print(prime_array)

prime_array_gt10 = prime_array > 10

prime_array_even = (prime_array % 2) != 1

print(prime_array_gt10)
print(prime_array_even)
```

### **3. Seleziona il numero più vicino a 0.5 in un array casuale**
Genera un array 10 x 3 con numeri casuali nell'intervallo [0,1].
Per ogni riga, seleziona il numero più vicino a 0.5. Per farlo usa il fancy indexing......

```python
import numpy as np

a = np.random.rand(10, 3)
a

distance = np.abs(a - 0.5)
distance

min_elements = np.argmin(distance, axis=1)
min_elements
index = np.arange(10)
index
closestlist = a[index, min_elements]
closestlist

print(a, distance, min_elements, index, closestlist)
```

### **4. Analisi della Frequenza Cardiaca con NumPy**

1. **Supponiamo che i seguenti valori rappresentino la tua frequenza cardiaca** registrata dal tuo Fitbit durante la giornata:  
   `68, 65, 77, 110, 160, 161, 162, 161, 160, 161, 162, 163, 164, 163, 162, 100, 90, 97, 72, 60, 70`  
   Inserisci questi valori in un array NumPy.

2. **Trova la frequenza cardiaca minima** registrata durante la giornata.  
   _(Questo valore rappresenta approssimativamente la tua frequenza cardiaca a riposo, un comune indicatore di salute.)_

3. **Trova la frequenza cardiaca massima** registrata durante la giornata.  
   _(Questo valore è un'indicazione dell'intensità dell'esercizio fisico.)_

4. **Calcoliamo la percentuale di letture effettuate mentre stavi facendo esercizio fisico (frequenza cardiaca sopra a 120)**:  
   - Crea un nuovo array che contiene `True` quando la frequenza cardiaca è **superiore a 120**, e `False` quando è **inferiore o uguale a 120**.
   - Utilizza una funzione di sintesi per calcolare la percentuale di osservazioni in cui la tua frequenza cardiaca era sopra 120.


```python
import numpy as np

# Creo l'array
frequency = np.array([68,65,77,110,160,161,162,161,160,161,162,163,164,163,162,100,90,97,72,60,70])

# Tramite funzione min e max trovo il valore più basso e più alto nell'array
minimum_frequency = int(np.min(frequency))
minimum_frequency
maximum_frequency = int(np.max(frequency))
maximum_frequency

# Con un confronto creo un array di true e false
stress_frequency = frequency > 120
stress_frequency

# Sfrutto la proprietà dei true e false ( 1 e 0 ) per trovare la media nell'array
# sommando tutti i true e dividendoli per la totalità degli elementi. Ricavo un numero fra 0 e 1 che
# Moltiplicato per 100 e arrotondato ( non ci interessano tutti i decimali, arrotondiamo al primo decimale
stress_percentage = round(float(np.mean(stress_frequency) * 100), 1)
stress_percentage
```

6. L'aumento dato al CEO e all'impiegata ha irritato gli altri due dipendenti, quindi ora devono ricevere un aumento del **10%**.

7. Calcola il **costo totale del personale** dopo tutti gli aumenti.  
Alla fine, quanto è costato all'azienda l'aumento di circa **16.000 dollari** del CEO?


```python
import numpy as np

salaries = np.array([50000,105250,55000,89000])
# Per vedere il totale degli stipendi utilizzo la funzione sum e forzo int così da ricevere solo il risultato
total_cost = int(np.sum(salaries))

# Do per scontato che il valore del CEO sia il valore più alto 
# ( devo o conoscere il valore dello stipendio o trovare un modo per dedurlo )
salaries[np.argmax(salaries)] = salaries[np.argmax(salaries)] * 1.15
floatedsalaries = salaries.astype(float)
floatedsalaries

# Il valore non è corretto avendo trasformato il valore prima, quindi ricreiamo il l'array questa volta già
# come un array di floats. Avendo modificato il mio dataset iniziale non posso più reperire il valore se non
# reinserendolo manualmente. Per comodità creo una copia dell'array di partenza
floatedsalaries, floatedsalariescopy = np.array([50000,105250,55000,89000], dtype=float), floatedsalaries.copy()
floatedsalaries[np.argmax(floatedsalaries)] = floatedsalaries[np.argmax(floatedsalaries)] * 1.15
print(floatedsalaries)

# Analogamente, gli altri aumenti e la somma totale si possono svolgere come prima
floatedsalaries[np.argmin(floatedsalaries)] = floatedsalaries[np.argmin(floatedsalaries)] * 1.20
print(floatedsalaries)
# Realisticamente, avendo aumentato il minimo e il massimo, ho reso la lista disordinata, non sapendo più
# chi ha ricevuto l'aumento. La confronto con la sua copia di partenza e vado ad aumentare le celle True

floatedsalaries_confront = floatedsalaries == floatedsalariescopy
floatedsalaries[floatedsalaries_confront] = floatedsalaries[floatedsalaries_confront] * 1.10
floatedsalaries
floated_total_cost = float(np.sum(floatedsalaries))
floated_total_cost
```
